# Hybrid (first->second order) optimization for PINNs

In this notebook we study **optimization strategies for Physics-Informed Neural Networks (PINNs)**,
focusing on a one-dimensional Poisson problem.

We compare:
- Adam,
- L-BFGS,
- a hybrid Adam → L-BFGS strategy with automatic switching.

Adam is an adaptive first-order optimization that maintains per-parameter adaptive learning rates via momentum and second moments. It is inherently robust to poor scaling and initialization and is thus excellent in early training when gradients may be unreliable. It may stagnate near convergence.

L-BFGS is a quasi-Newton method that
builds a low-rank approximation of the inverse Hessian from curvature history. It
achieves superlinear convergence near a minimizer and is thus
very effective for final accuracy improvements. It is better suited to the second stage of the optimization, since it requires good initialization since poor starting points can lead to failures.


The key insight of the hybrid strategy is to combine the complementary strengths of the two methods:
- **Adam phase** (exploration): robust initialization and coarse convergence to a good region
- **L-BFGS phase** (refinement): fast superlinear convergence to high accuracy



```python
import torch
import torch.nn as nn
import torch.optim as optim
import numpy as np
import matplotlib.pyplot as plt

torch.manual_seed(0)
np.random.seed(0)
```

## 2. Model problem: 1D Poisson equation

We consider the Poisson problem on $\Omega = (0,1)$:
$$-u''(x) = f(x), \qquad u(0)=u(1)=0,$$
with a forcing term containing multiple frequencies.

```python
def u_exact_np(x):
    """Manufactured exact solution of the 1D Poisson problem.

    u(x) = sin(pi x) + 0.3 sin(4 pi x) + 0.1 sin(9 pi x).
    The multi-frequency content makes the problem moderately stiff
    for first-order optimizers.

    Parameters
    ----------
    x : float or ndarray
        Point(s) in [0, 1].

    Returns
    -------
    u : same shape as x
        Exact solution evaluated at x.
    """
    return (
        np.sin(np.pi * x)
        + 0.3 * np.sin(4 * np.pi * x)
        + 0.1 * np.sin(9 * np.pi * x)
    )

def f_force(x):
    """Forcing term f(x) = -u''(x) consistent with `u_exact_np`.

    Differentiating u_exact twice and negating gives this trigonometric
    sum. Implemented in torch so it can be used with torch tensors of
    collocation points.

    Parameters
    ----------
    x : torch.Tensor
        Collocation points (any shape).

    Returns
    -------
    f : torch.Tensor of the same shape as x
        Right-hand side of the PDE.
    """
    return (
        (np.pi**2) * torch.sin(np.pi * x)
        + 0.3 * (4*np.pi)**2 * torch.sin(4*np.pi * x)
        + 0.1 * (9*np.pi)**2 * torch.sin(9*np.pi * x)
    )
```

## 3. PINN with embedded boundary conditions and collocation points

We parameterize the solution as:
$$u_\theta(x) = (1-x)u(0) + x u(1) + x(1-x)\,h_\theta(x),$$
where $h_\theta(x)$ is a neural network. In this way we enforce boundary conditions by embedding them into the network
architecture. This approach guarantees that BC are exactly satisfied and
eliminates the weight hyperparameters that require tuning.

We sample the collocation points uniformly in $[0,1]$.

```python
# Extract boundary values
u0 = float(u_exact_np(0.0))
u1 = float(u_exact_np(1.0))

# Training mesh
Ntrain = 300
x_np = np.linspace(0.0, 1.0, Ntrain)
x = torch.tensor(x_np, dtype=torch.float32).unsqueeze(1)


# Reference solution and forcing function f_rhs = f_force(x)
u_exact = torch.tensor(u_exact_np(x_np), dtype=torch.float32).unsqueeze(1)
f_rhs = f_force(x)

class PINN(nn.Module):
    """Physics-Informed Neural Network with hard-encoded Dirichlet BCs.

    The output is decomposed as
        u_theta(x) = (1 - x) * u(0) + x * u(1) + x * (1 - x) * h_theta(x),
    where h_theta is a small fully-connected tanh network. The factor
    x*(1-x) vanishes at the endpoints, so the boundary conditions are
    satisfied exactly by construction; only the interior PDE residual
    needs to be optimized.
    """
    def __init__(self, width=128):
        """Build a 1 -> width -> width -> 1 tanh MLP.

        Parameters
        ----------
        width : int
            Hidden width of the two interior layers.
        """
        super().__init__()
        self.net = nn.Sequential(
            nn.Linear(1, width), nn.Tanh(),
            nn.Linear(width, width), nn.Tanh(),
            nn.Linear(width, 1)
        )

    def forward(self, x):
        """Evaluate the PINN solution at points x with embedded BCs.

        Parameters
        ----------
        x : torch.Tensor of shape (N, 1)
            Collocation or evaluation points.

        Returns
        -------
        u : torch.Tensor of shape (N, 1)
            Approximate solution values, exactly matching u(0)=u0,
            u(1)=u1 by construction.
        """
        n = self.net(x)
        # boundary embeddings
        return (1 - x) * u0 + x * u1 + x*(1-x) * n

```

## 4. Computing the PDE residual

The Poisson equation involves second order derivatives, which we compute
using automatic differentiation, and which will be used to copute the PINN loss.

```python
def pinn_second_derivative(model, x):
    """Second derivative d2 u_theta / dx2 by repeated automatic differentiation.

    Builds a fresh leaf tensor xg with `requires_grad=True`, evaluates
    the model, and applies torch.autograd.grad twice with
    `create_graph=True` so the result remains differentiable
    (necessary because this enters the loss whose gradient w.r.t.
    parameters we still need to compute).

    Parameters
    ----------
    model : torch.nn.Module
        Network mapping x -> u_theta(x).
    x : torch.Tensor of shape (N, 1)
        Collocation points (does not need to require grad; cloned
        internally).

    Returns
    -------
    d2u : torch.Tensor of shape (N, 1)
        Second derivative evaluated at each x_i, attached to the
        autograd graph w.r.t. model parameters.
    """
    xg = x.clone().detach().requires_grad_(True)
    u = model(xg)
    # First derivative
    du = torch.autograd.grad(u, xg, torch.ones_like(u), create_graph=True)[0]
    # Second derivative
    d2u = torch.autograd.grad(du, xg, torch.ones_like(du), create_graph=True)[0]

    return d2u

def loss_fn(uxx, f_rhs):
    """PDE-residual loss for -u''(x) = f(x).

    Returns the mean of (u'' + f)^2 over collocation points. Since
    -u'' = f is equivalent to u'' + f = 0, minimizing this mean square
    drives the PDE residual to zero.

    Parameters
    ----------
    uxx : torch.Tensor of shape (N, 1)
        Second derivative of the PINN at the collocation points.
    f_rhs : torch.Tensor of shape (N, 1)
        Forcing values at the same collocation points.

    Returns
    -------
    loss : 0-d torch.Tensor
        Scalar mean-squared residual.
    """
    return ((uxx + f_rhs) ** 2).mean()

```

## 5. Utility functions

We use helper routines to clone identical initializations and to evaluate the PDE loss and the solution error.

```python
def clone_params_from(model_source, model_target):
    """Copy parameter tensors from `model_source` into `model_target` in place.

    Both models must have the same parameter shapes and ordering
    (same architecture). Used to ensure the optimizer comparisons
    start from identical initializations.

    Parameters
    ----------
    model_source : torch.nn.Module
        Model whose parameter values are copied.
    model_target : torch.nn.Module
        Model whose parameter values are overwritten in-place.
    """
    with torch.no_grad():
        for p_src, p_tgt in zip(model_source.parameters(), model_target.parameters()):
            p_tgt.copy_(p_src)

def compute_loss_and_error(model, x, f_rhs, u_exact):
    """Evaluate the PDE residual loss and the solution L2 error.

    Parameters
    ----------
    model : torch.nn.Module
        Trained or partially trained PINN.
    x : torch.Tensor of shape (N, 1)
        Collocation/evaluation points.
    f_rhs : torch.Tensor of shape (N, 1)
        Forcing values at those points.
    u_exact : torch.Tensor of shape (N, 1)
        Reference solution values (used only for diagnostics).

    Returns
    -------
    loss : float
        Mean-squared PDE residual.
    err : float
        Euclidean norm of (u_theta(x) - u_exact(x)).
    """
    uxx = pinn_second_derivative(model, x)
    loss = loss_fn(uxx, f_rhs).item()
    err = torch.norm(model(x) - u_exact).item()
    return loss, err
```

## 6. Optimizers:  Adam and L-BFGS

We use PyTorch's optimizers. PyTorch's L-BFGS requires a *closure* because it may evaluate
the loss and gradient multiple times per step.



```python
def train_adam(model, x, f_rhs, u_exact, iters=2000, lr=1e-3):
    """Train a PINN with Adam for a fixed number of iterations.

    Each iteration performs a single full-batch update on the PDE
    residual loss mean((u_theta'' + f)^2). Loss and L2 error are
    recorded after every step for plotting.

    Parameters
    ----------
    model : torch.nn.Module
        PINN to be trained (parameters updated in place).
    x : torch.Tensor of shape (N, 1)
        Collocation points.
    f_rhs : torch.Tensor of shape (N, 1)
        Forcing values at the collocation points.
    u_exact : torch.Tensor of shape (N, 1)
        Reference solution (only for monitoring).
    iters : int
        Number of Adam iterations.
    lr : float
        Adam learning rate.

    Returns
    -------
    losses : ndarray of shape (iters,)
        Per-iteration PDE residual loss.
    errs : ndarray of shape (iters,)
        Per-iteration solution error.
    """
    opt = optim.Adam(model.parameters(), lr=lr)
    losses, errs = [], []

    for k in range(iters):
        opt.zero_grad()
        uxx = pinn_second_derivative(model, x)
        loss = loss_fn(uxx, f_rhs)  # PDE residual: ||u'' + f||^2 (== ||-u'' - f||^2)
        loss.backward()
        opt.step()

        L, E = compute_loss_and_error(model, x, f_rhs, u_exact)
        losses.append(L)
        errs.append(E)

        if k % 200 == 0:
            print(f"[Adam] iter {k:4d} | loss={L:.3e} | err={E:.3e}")

    return np.array(losses), np.array(errs)
```

```python
def train_lbfgs(model, x, f_rhs, u_exact,
                max_iters=500,
                history_size=50,
                tol_grad=1e-12,
                tol_change=1e-12):
    """Train a PINN with L-BFGS (strong-Wolfe line search).

    Each outer iteration calls `optim.LBFGS.step(closure)` with
    `max_iter=1`, which performs a single quasi-Newton update using
    the supplied closure to recompute loss and gradient. A simple
    early-stop heuristic terminates training if the loss stays
    essentially unchanged for 20 consecutive iterations.

    Parameters
    ----------
    model : torch.nn.Module
        PINN to be trained (in place).
    x : torch.Tensor of shape (N, 1)
        Collocation points.
    f_rhs : torch.Tensor of shape (N, 1)
        Forcing values.
    u_exact : torch.Tensor of shape (N, 1)
        Reference solution (monitoring only).
    max_iters : int
        Maximum number of outer L-BFGS iterations.
    history_size : int
        L-BFGS curvature memory length.
    tol_grad : float
        Optimizer's gradient tolerance.
    tol_change : float
        Optimizer's tolerance on parameter / loss change.

    Returns
    -------
    losses, errs : ndarray
        Per-iteration loss and error history (truncated to the actual
        number of iterations performed).
    """
    losses, errs = [], []

    opt = optim.LBFGS(
        model.parameters(),
        max_iter=1,
        history_size=history_size,
        line_search_fn="strong_wolfe",
        tolerance_grad=tol_grad,
        tolerance_change=tol_change
    )

    prev_L = float('inf')
    stale = 0
    for k in range(max_iters):
        # L-BFGS requires a closure for re-evaluation
        def closure():
            opt.zero_grad()
            uxx = pinn_second_derivative(model, x)
            loss = loss_fn(uxx, f_rhs)
            loss.backward()
            return loss

        opt.step(closure)

        L, E = compute_loss_and_error(model, x, f_rhs, u_exact)
        losses.append(L)
        errs.append(E)

        if k % 50 == 0:
            print(f"[LBFGS] iter {k:4d} | loss={L:.3e} | err={E:.3e}")

        # Early stop: if loss hasn't improved relatively for a while, break out.
        if abs(prev_L - L) <= 1e-10 * max(1.0, abs(L)):
            stale += 1
            if stale >= 20:
                print(f"[LBFGS] stalled at iter {k} (loss={L:.3e}); stopping early.")
                break
        else:
            stale = 0
        prev_L = L

    return np.array(losses), np.array(errs)
```

## 7. Automatic Adam → L-BFGS switching

The key point to combine Adam and L-BFGS is to understand when to switch from one to the other. In this experiment we monitor the gradient norm $\|\nabla f\|_2$ during Adam training.
When its decrease stagnates over a sliding window, we switch automatically to L-BFGS.

More precisely, we compare the mean gradient norm over recent iterations versus earlier iterations.
If the ratio exceeds a threshold (here 0.995), convergence has plateaued and it's time to switch.

This provides a principled, data-driven transition criterion that requires no manual tuning
of iteration counts or wall-clock time.

```python
def train_adam_then_lbfgs_automatic_switch(
        model,
        x, f_rhs, u_exact,
        adam_max_iters=8000,
        adam_min_iters=500,
        lbfgs_iters=1000,
        lr_adam=1e-4,
        window=200,
        plateau_tol=0.995,
        history_size=50,
        tol_grad=1e-12,
        tol_change=1e-12):
    """Hybrid Adam -> L-BFGS training with automatic plateau-based switching.

    Phase 1 runs Adam and tracks the gradient L2-norm at every step.
    After at least `adam_min_iters` iterations, it compares the mean
    gradient norm over the last `window` steps to the mean over the
    preceding `window` steps. If the ratio exceeds `plateau_tol`
    (default 0.995, i.e. < 0.5% improvement), Adam is judged to be
    stalling and training switches to phase 2.

    Phase 2 runs L-BFGS (strong-Wolfe line search) until either
    `lbfgs_iters` outer iterations have elapsed or the loss has been
    essentially unchanged for 20 consecutive iterations.

    Parameters
    ----------
    model : torch.nn.Module
        PINN to be trained (in place).
    x, f_rhs, u_exact : torch.Tensor
        Collocation points, forcing, and reference solution.
    adam_max_iters : int
        Hard cap on the Adam phase.
    adam_min_iters : int
        Minimum Adam iterations before plateau detection is allowed.
    lbfgs_iters : int
        Hard cap on the L-BFGS phase.
    lr_adam : float
        Adam learning rate.
    window : int
        Sliding-window size used by the plateau detector.
    plateau_tol : float
        Ratio threshold above which the gradient norm is considered
        flat (default 0.995).
    history_size, tol_grad, tol_change : see `train_lbfgs`.

    Returns
    -------
    losses, errs : ndarray
        Combined loss and error history across both phases.
    grad_norms : ndarray
        Per-iteration gradient norm during the Adam phase (used for
        plateau diagnostics).
    """
    losses, errs, grad_norms = [], [], []

    opt_adam = optim.Adam(model.parameters(), lr=lr_adam)

    def grad_norm():
        """L2 norm of the concatenated parameter gradients of `model`."""
        return sum(p.grad.norm().item()**2 for p in model.parameters()
                   if p.grad is not None)**0.5

    # ----- Adam stage: early exploration -----
    for k in range(adam_max_iters):
        opt_adam.zero_grad()
        uxx = pinn_second_derivative(model, x)
        loss = loss_fn(uxx, f_rhs)
        loss.backward()
        opt_adam.step()

        L, E = compute_loss_and_error(model, x, f_rhs, u_exact)
        gnorm = grad_norm()

        losses.append(L)
        errs.append(E)
        grad_norms.append(gnorm)

        # Check for convergence plateau using sliding window
        if k > adam_min_iters + window:
            r = np.mean(grad_norms[-window:]) / np.mean(grad_norms[-2*window:-window])
            if r > plateau_tol:  # Stagnation detected
                print(f">>> Switching to L-BFGS at iter {k}")
                break

    # ----- L-BFGS stage: final refinement -----
    opt_lbfgs = optim.LBFGS(
        model.parameters(),
        max_iter=1,
        history_size=history_size,
        line_search_fn="strong_wolfe",
        tolerance_grad=tol_grad,
        tolerance_change=tol_change
    )

    prev_L = float('inf')
    stale = 0
    for k in range(lbfgs_iters):
        def closure():
            opt_lbfgs.zero_grad()
            uxx = pinn_second_derivative(model, x)
            loss = loss_fn(uxx, f_rhs)
            loss.backward()
            return loss

        opt_lbfgs.step(closure)

        L, E = compute_loss_and_error(model, x, f_rhs, u_exact)
        losses.append(L)
        errs.append(E)

        if abs(prev_L - L) <= 1e-10 * max(1.0, abs(L)):
            stale += 1
            if stale >= 20:
                print(f"[Hybrid/LBFGS] stalled at iter {k} (loss={L:.3e}); stopping early.")
                break
        else:
            stale = 0
        prev_L = L

    return np.array(losses), np.array(errs), np.array(grad_norms)
```

## 8. Comparison

We compare the three optimizers starting from the same initial guess.

```python
# Initialize three identical PINNs
model_adam = PINN()
model_lbfgs = PINN()
model_hybrid = PINN()

# Clone parameters for fair comparison
clone_params_from(model_adam, model_lbfgs)
clone_params_from(model_adam, model_hybrid)

# Train each model with different strategies
loss_adam, err_adam = train_adam(model_adam, x, f_rhs, u_exact, iters=5000, lr=5e-4)
loss_lbfgs, err_lbfgs = train_lbfgs(model_lbfgs, x, f_rhs, u_exact, max_iters=5000)
loss_hybrid, err_hybrid, _ = train_adam_then_lbfgs_automatic_switch(model_hybrid,
                                                                    x, f_rhs, u_exact,
                                                                    adam_max_iters=8000,
                                                                    adam_min_iters=500,
                                                                    lbfgs_iters=1500,
                                                                    lr_adam=5e-4,)
```

```python
# Compare convergence curves
plt.figure(figsize=(12,5))

# PDE residual loss
plt.subplot(1,2,1)
plt.semilogy(loss_adam, label="Adam")
plt.semilogy(loss_lbfgs, label="L-BFGS")
plt.semilogy(loss_hybrid, label="Adam → L-BFGS")
plt.title("PINN PDE Loss")
plt.legend(); plt.grid(True)

# Solution error
plt.subplot(1,2,2)
plt.semilogy(err_adam, label="Adam")
plt.semilogy(err_lbfgs, label="L-BFGS")
plt.semilogy(err_hybrid, label="Adam → L-BFGS")
plt.title("PINN Solution Error")
plt.legend(); plt.grid(True)

plt.tight_layout()
plt.show()
```

## 9. Discussion

We can see that the hybrid strategy outperforms both Adam and L-BFGS. Even if L-BFGS provides a quicker initial decrease, it later stagnates.
The inital phase with Adam avoids L-BFGS to get stack, by providing a better starting point.

The automatic switching strategy uses gradient norm stagnation as a signal, eliminating manual iteration count guessing. This approach scales well to high-dimensional PINNs and other PDE-constrained optimization problems.
